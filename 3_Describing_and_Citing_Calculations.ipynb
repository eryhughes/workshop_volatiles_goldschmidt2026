# 3. Describing and citing calculations

It is important to properly cite all models, tools, and frameworks used for calculations. Here is an example of how you might describe and cite the calculations presented in the previous notebooks - including the previous notebooks in Supplementary Material would ensure the calculations are reproduceable!

All calculations were performed using the VICTOR platform (Lev et al., 2025). Melt inclusion data is from Wieser et al. (2021) and the fluid inclusion data from DiVitre & Wieser (2024). Temperature from melt melt inclusion data was calculated using the MgO-liquid thermometer from Helz & Thornber (1987) implemented in Thermobar (v1.0.73; Wieser et al., 2022). Pressure and fluid composition for melt inclusions, and degassing paths, were calculated using the model of Iacono-Marziano et al. (2012) implemented in VESIcal (v1.2.10; Iacovino et al., 2021) and models shown in Table 1 implemented in VolFe (v1.0.2; Hughes et al., 2025). Pressures from fluids inclusions were calculated using the equation of state from Sterner and Pitzer (1994) implemented in DiadFit (v1.0.23, Wieser et al., 2024). Degassing paths were additionally calculated using Sulfur_X (v1.2; Ding et al., 2023), which includes the models of Iacono-Marziano et al. (2012) and O'Neill & Mavrogenes (2022), and EVo (Liggins et al., 2020, 2022).
