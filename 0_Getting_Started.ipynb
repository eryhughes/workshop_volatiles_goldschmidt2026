If you are running this notebook through your computer, you will need to install the relevant packages. Follow these instructions to set up your computer's environment <em>before</em> running any of the example notebooks. We have provided an ``environment.yaml`` file in the root directory that will automatically build a conda environment for you and install the dependencies needed to run these notebooks. To run it, open a terminal (Mac/Linux) or PowerShell, Command Prompt, or Anaconda Prompt (Windows), change directory `cd` to the folder where this notebook lives, then type:

```sh
conda env create -f environment.yaml
conda activate volatiles-gs-workshop26
```

**Note**: If running in a jupyter environment, make sure you select volatiles-gs-workshop26 as your kernel.
